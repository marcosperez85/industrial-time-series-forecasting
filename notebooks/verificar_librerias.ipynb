```python
# Ejecuta esto en una celda del notebook para verificar dependencias
import sys

required_modules = [
    'yaml', 'pandas', 'numpy', 'matplotlib', 
    'seaborn', 'sklearn', 'joblib'
]

missing_modules = []
for module in required_modules:
    try:
        __import__(module)
        print(f"‚úÖ {module}")
    except ImportError:
        print(f"‚ùå {module} - FALTA")
        missing_modules.append(module)

if missing_modules:
    print(f"\nüì¶ Instalar faltantes:")
    if 'yaml' in missing_modules:
        print("pip install pyyaml")
    for module in missing_modules:
        if module != 'yaml':
            print(f"pip install {module}")
else:
    print(f"\n‚úÖ Todas las dependencias est√°n instaladas!")
```
